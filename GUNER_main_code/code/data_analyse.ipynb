```python
import json
len_dict = {}
entity_label = {}
with open("/home/jw/CCL_Guner2023/data/train_json/all_train.json",'r') as f:
    f = json.load(f)
    for lines in f:
        for sub_list in lines['entity_list']:
            if sub_list[-1] not in entity_label.keys():
                entity_label[sub_list[-1]] = 1
            else:
                entity_label[sub_list[-1]] += 1
        if len(lines['text']) not in len_dict.keys():
            len_dict[len(lines['text'])] = 1
        else:
            len_dict[len(lines['text'])] += 1
```
